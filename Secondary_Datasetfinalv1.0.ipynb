##### If average cohen kappa score for any annotation is less than 0.2, we will drop it

```python
def drop_annotate():
    average = [cohen_avg_96(), cohen_avg_53(),cohen_avg_93(),cohen_avg_94(), cohen_avg_95()]
    columns_name = [['annotation_96'],['annotation_53'],['annotation_93'],['annotation_94'],['annotation_95']]
    for i in range(0,len(average)):
        if average[i]< 0.2:
            print(columns_name[i])
            print(average[i])
            df.drop(columns_name[i], axis=1, inplace = True)
    print(df.head())
    r = []
    for col in df.columns:
        r.append(col)
    print(r[1:])
```

```python
drop_annotate()
```

```python
def finalLabel(df):
    for index, rows in df.iterrows():
        #print(index, rows['text'])
        compare = [[cohen_avg_96(),rows['annotation_96']],
                       [cohen_avg_93(),rows['annotation_93']],
                   [cohen_avg_94(),rows['annotation_94']],
                   [cohen_avg_95(),rows['annotation_95']]]
        
        boolean = [rows['annotation_96'],rows['annotation_93'],rows['annotation_94'],rows['annotation_95']]
        
        compare.sort(key = lambda k:k[0])
        t = compare.count('True')
        f = compare.count('False')
        #print('True:',t, 'False', f)
        
        if t == f: 
            kapaValue, colValue = compare[-1]
            df.loc[index, 'final_label'] = colValue
            
        elif t!=f:
            if t > f:
                df.loc[index, 'final_label'] = True
            else:
                df.loc[index, 'final_label'] = False
    
finalLabel(df)
df = df.iloc[:,[0,-1]]
df.head()
```
